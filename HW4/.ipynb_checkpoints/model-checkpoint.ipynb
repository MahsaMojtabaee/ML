```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('data\d1.csv')
data1 = np.array(df1)
```

```python
df1.describe()
```

```python
plt.scatter(df1.iloc[:, 0], df1.iloc[:, 1])
```

# First Dataset

```python
eps_s = [0.001, 0.005, 0.01, 0.02, 0.1, 0.5, 1]
mins_s = [3, 4, 5, 6]

# Create subplots
fig, axes = plt.subplots(len(eps_s), len(mins_s), figsize=(12, 8), sharex=True, sharey=True)
plt.subplots_adjust(top=1.5)

for i, ep in enumerate(eps_s):
    for j, mns in enumerate(mins_s):
        clf = DBSCAN(eps=ep, min_samples=mns)
        clf.fit(data1)
        labels = clf.get_labels()
        sns.scatterplot(x=data1[:, 0], y=data1[:, 1], hue=labels, ax=axes[i, j], palette='viridis', s=50)
        axes[i, j].set_title(f'Eps={ep}, Min Samples={mns}')    
    
```
