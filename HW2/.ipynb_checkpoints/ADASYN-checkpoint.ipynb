```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import ADASYN
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn import svm
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
```

```python
data = pd.read_csv('data/dataset.csv')
```

```python
data.info()
```

from here we can see there are missing values in smoking_status and bmi

I choose to check what to do with these missed values. Here my options: 
1. Drop those people from my training set 
2. Replace the Nan with the mean of the column
3. Replace the Nan with the max of the column
4. Replace the Nan with the most frequent value of the column

```python
target = data.stroke

replacement_map = {
    'ever_married':{'Yes':1, 'No':0},
    'Residence_type':{'Urban':1, 'Rural':0}
}


first_x = data.drop(['stroke', 'smoking_status', 'bmi'], axis=1).replace(replacement_map)
x = pd.get_dummies(first_x, columns=['work_type', 'gender'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, target, test_size=0.3, random_state=0, stratify=target)
print(f'shape of train dataset is x:{x_train.shape} and y:{y_train.shape}')
print(f'shape of test dataset is x:{x_test.shape} and y:{y_test.shape}')
```

```python
(target.value_counts(normalize=True)*100).plot(kind='bar', color='c')
plt.xlabel('Stroke')
plt.ylabel('Percentage')
plt.title('Percentage of Strokes')
```

```python
(y_train.value_counts(normalize=True)*100).plot(kind='bar', color='c')
plt.xlabel('Stroke')
plt.ylabel('Percentage')
plt.title('Train Set Percentage of Strokes')
```

```python
(y_test.value_counts(normalize=True)*100).plot(kind='bar', color='c')
plt.xlabel('Stroke')
plt.ylabel('Percentage')
plt.title('Test Set of Percentage of Strokes')
```

### We can see that the data is imbalance

Let's use the imbalnced data technique: in this model I use ADASYN

```python
x_resampled, y_resampled = ADASYN().fit_resample(x_train, y_train)
```

```python
print(x_resampled.shape)
print(y_resampled.shape)
```

```python
(y_resampled.value_counts(normalize=True)*100).plot(kind='bar', color='c')
plt.xlabel('Stroke')
plt.ylabel('Percentage')
plt.title('Percentage of Strokes')
```

Let's balance the data with normalization

```python
x_normalized = Normalizer().fit_transform(x_resampled)
columns = x_resampled.columns  # Use the feature names as columns for the DataFrame
x_normalized_df = pd.DataFrame(x_normalized, columns=columns)
x_normalized_df
```

```python
clf = svm.SVC(class_weight={1:100})
clf.fit(x_normalized_df, y_resampled)
```

```python
pred = clf.predict(x_test)
```

```python
confusion_matrix = confusion_matrix(y_test, pred)
sns.heatmap(confusion_matrix, annot=True)
```

```python
accuracy_score(y_test, pred)
```
